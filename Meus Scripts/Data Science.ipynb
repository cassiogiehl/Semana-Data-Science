```python
df.drop('vehicle.age.group', axis=1, inplace=True)
df.drop('data.set', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder1 = LabelEncoder()
label_encoder2 = LabelEncoder()
label_encoder3 = LabelEncoder()
label_encoder4 = LabelEncoder()
label_encoder5 = LabelEncoder()
label_encoder6 = LabelEncoder()
```

```python
df['vehicle.type'] = label_encoder1.fit_transform(df['vehicle.type'])
df['domestic.import'] = label_encoder2.fit_transform(df['domestic.import'])
df['color.set'] = label_encoder2.fit_transform(df['color.set'])
df['makex'] = label_encoder2.fit_transform(df['makex'])
df['state'] = label_encoder2.fit_transform(df['state'])
df['make.model'] = label_encoder2.fit_transform(df['make.model'])
```

```python
df.head()
```

```python
y = df['overage']
```

```python
y.head()
```

```python
# Apaga a coluna overage e lot.sale.days
X = df.drop(['lot.sale.days', 'overage'], axis=1)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# definindo variáveis de treino e teste
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y)
```

```python
X_treino.count()
```

```python
y_treino.count()
```

```python
X_treino.head()
```

```python
y_treino.head()
```

```python
X_teste.count()
```

```python
y_teste.count()
```

```python
X_teste.head()
```

```python
from sklearn import tree
```

```python
arvore = tree.DecisionTreeClassifier()
```

```python
arvore.fit(X_treino, y_treino)
```

```python
X_teste.head()
```

```python
arvore.predict(X_teste)
```

```python
resultado = arvore.predict(X_teste)
```

```python
resultado
```

```python
# Relatório de métricas
from sklearn import metrics
print(metrics.classification_report(y_teste, resultado))
```

```python
arvore.feature_importances_
```

```python
# Importância de cada atributo
feature_imp = pd.Series(arvore.feature_importances_, index=X_treino.columns)
```

```python
feature_imp
```

```python
# Plotando a importância de cada atributo (feature)
%matplotlib notebook
sns.barplot(x=feature_imp, y=feature_imp.index)

plt.xlabel('Importância de Features')
plt.ylabel('Features')
plt.title("Importância de Features")
plt.show()
```
